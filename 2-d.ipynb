```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l=4
```

```python
def isValid(s,a):


    return (s+a>=0 and s+a<l)
def tau(s,a):

    if(a==0):#up
        if(isValid(s[0],-1)):
            return (s[0]-1,s[1])
    elif(a==1):#down
        if(isValid(s[0],+1)):
            return (s[0]+1,s[1])
    elif(a==2):#left
        if(isValid(s[1],-1)):
            return (s[0],s[1]-1)    
    elif(a==3):#right
        if(isValid(s[1],1)):
            return (s[0],s[1]+1)    
    return(s)

def calc_policy(Q):
    pi = np.full((l, l), 0)
    for i in range(0,l):
        for j in range(0,l):
            pos=idq((i,j))
            action=np.argmax(Q[pos,:])
            pi[i,j]=action
            if(pos==5 or pos==13 or pos==15):
                pi[i,j]=-1

    return pi.astype(int)  

def idq(s):

    return ((s[0])*l)+s[1]
```

```python


Q = np.full((l*l, 4), 0.5)
Q[5,:]=-1
Q[13,:]=-1
Q[15,:]=1
Q
```

```python
gamma = 0.9
initial_state = [0,0]
```

```python
def update(current_state, action, gamma): 
  
    next_state = tau(current_state, action)



    current_statePos=idq(current_state)
    next_statePos=idq(next_state)
    Q[current_statePos, action] = Q[current_statePos, action] - gamma * (Q[current_statePos, action]- np.average(Q[next_statePos, :]))
    return next_state
```

```python

def learning_function(Q, gamma):

    current_state = initial_state
    pos=idq(current_state)  


    i=0

    

    while pos != 5 and pos != 13 and pos != 15 :
        i+=1
        print("iteration: ",i)
        if(i%2==0):
            print("r")
            action = np.random.choice([0, 1,2,3])
        else:  
            print("m")  

            action = np.argmax(Q[pos, :])

        current_state = update(current_state, action, gamma)
        pos=idq(current_state) 
        print("current state: ",current_state,pos,(pos != 5 or pos != 13 or pos != 15))
    return Q
```

```python
for i in range(100):
    Q = learning_function(Q, gamma)
    
```
